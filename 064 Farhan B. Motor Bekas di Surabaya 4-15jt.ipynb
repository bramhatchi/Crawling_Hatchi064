```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_autoinstaller
import csv
import time
```

```python
#INSTALL BROWSER SUCH CHROME OR EDGE
chromedriver_autoinstaller.install()

#USE THE INSTALLED BROWSER FOR SELENIUM
browser = webdriver.Chrome()
```

```python
data = []

for index in range (1,3):
    browser.get(f'https://www.olx.co.id/surabaya-kota_g4000216/q-motor-bekas?filter=price_between_4000000_to_15000000')
    time.sleep(20)
    soup = BeautifulSoup(browser.page_source,"lxml")
    rows = soup.find_all('div', class_='fTZT3')
```

```python
    for row in rows:
        MerkdanTipeMotor = row.find(class_='_2poNJ')
        if MerkdanTipeMotor:
            MerkdanTipeMotor = MerkdanTipeMotor.text
            
        TahunMotor = row.find(class_='YBbhy')
        if TahunMotor:
            TahunMotor = TahunMotor.text
            
        LokasiPenjual = row.find(class_='_2VQu4')
        if LokasiPenjual:
            LokasiPenjual = LokasiPenjual.text
            
        Harga = row.find(class_='_2Ks63')
        if Harga:
            Harga = Harga.text
        
        WaktuPenjualan = row.find(class_='_2jcGx')
        if WaktuPenjualan:
            WaktuPenjualan = WaktuPenjualan.text
        
        df = {
            'Merk dan Tipe Motor': MerkdanTipeMotor,
            'Tahun Motor': TahunMotor,
            'Lokasi Penjual': LokasiPenjual,
            'Harga': Harga,
            'WaktuPenjualan': WaktuPenjualan
            }
        data.append(df)
```

```python
df = pd.DataFrame(data)
```
